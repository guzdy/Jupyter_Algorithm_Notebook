## 1. 冒泡排序 Bubble Sort

```python
def bubble_sort(lst):
    length = len(lst) 
    for i in range(length):
        swapped = False # 确定是否完成。如果无变更，则停止循环
        for j in range(0,length-1-i): # 每循环一次，后面增加确认好的队伍。
            if lst[j] > lst[j+1]: # 前面的元素比后面的元素大，换位置
                lst[j],lst[j+1] = lst[j+1],lst[j]
                swapped = True # 有变更。
        if not swapped:
            break
    return lst

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
bubble_sort(lst)
```

## 2. 选择排序 Selection Sort

```python
def slct_sort(lst):
    length = len(lst)
    for i in range(length):
        min_index = i
        for j in range(i+1,length): # 每循环一次，前面就有一个元素就确认好
            if lst[min_index] > lst[j]:
                min_index = j
        lst[i],lst[min_index] = lst[min_index],lst[i]
    return lst

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
slct_sort(lst)
```

## 3. 插入排序

```python
def insertion_sort(lst):
    length = len(lst)
    for i in range(1, length): # 循环比较从第二个数开始，最后一个数结束
        insertion_digit = lst.pop(i) # 拿出来选择插入数据
        for j in range(i-1,-1,-1): # 从选择数据前面一个开始比较到 0 
            if lst[j] < insertion_digit: # 如果前面的数小于拿出来的数
                lst.insert(j+1, insertion_digit) # 插入到刚才数后面
                break
        else: # 如果未找到更小的树，插入到最前面
            lst.insert(0, insertion_digit)
    return lst

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
insertion_sort(lst)
```

## 4. 希尔排序

```python
def shell_sort(lst): # 1/2 步长希尔函数
    length = len(lst)
    gap = length // 2
    while gap > 0:
        for start in range(gap): # 根据起始位置分表
            _insertion_sort(lst, start, gap, length)
        #print("After increments of size",gap, "The list is",lst)
        gap = gap // 2
        
    return lst

def _insertion_sort(lst, start, gap, length): 
    # 因为需要按跨度推后数据，所以 python 插入法只能交换数据
    for i in range(start+gap,length,gap): # 选择某列中的数据
        current_value = lst[i]
        position = i
        
        while position >= gap and lst[i] > current_value:
            lst[positon] = lst[positon-gap]
            positon = positon-gap
        
        lst[position]=current_value
        
    return lst

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
shell_sort(lst)    
```

## 5. 堆排序

```python
def heap_sort(ary) :
    n = len(ary)
    first = int(n/2-1) #最后一个非叶子节点
    for start in range(first,-1,-1) : #构造大根堆
        max_heapify(ary,start,n-1)
    for end in range(n-1,0,-1):   #堆排，将大根堆转换成有序数组
        ary[end],ary[0] = ary[0],ary[end]
        max_heapify(ary,0,end-1)
    return ary

#最大堆调整：将堆的末端子节点作调整，使得子节点永远小于父节点
#start为当前需要调整最大堆的位置，end为调整边界
def max_heapify(ary,start,end):
    root = start
    while True :
        child = root*2 +1               #调整节点的子节点
        if child > end : break
        if child+1 <= end and ary[child] < ary[child+1] :
            child = child+1             #取较大的子节点
        if ary[root] < ary[child] :     #较大的子节点成为父节点
            ary[root],ary[child] = ary[child],ary[root]     #交换
            root = child
        else : break

            
lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
heap_sort(lst)
```

```python
import heapq

def heapsort(iterable):
    h = []
    for value in iterable:
        heapq.heappush(h,value)
    # heappush(heap,item) 往堆中插入一条新的值 
    # heappop(heap) 从堆中弹出最小值 
    return [heapq.heappop(h) for i in range(len(h))] 

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
heapsort(lst) 
```

```python
import heapq

def HeapSort(lst):
    heapq.heapify(lst) # 对于已有的列表转换成堆
    return [heapq.heappop(lst) for i in range(len(lst))] 

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
HeapSort(lst)
```

## 归并排序

```python
def merge_sort(lst):
    length = len(lst)
    if length <= 1: return lst
    mid = length // 2  # 二分分解
    left = merge_sort(lst[:mid])
    right = merge_sort(lst[mid:])
    return merge(left, right)
    
def merge(left, right):
    result = []
    while len(left) > 0 and len(right) > 0:
        if left[0] > right[0]:
            result.append(right.pop(0))
        else:
            result.append(left.pop(0))
    result += right
    result += left
    return result

lst = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
merge_sort(lst)
```

## 快速排序

```python
def quick_sort(array, begin=0, end=None):
    if end is None:
        end = len(array) - 1
    def _quicksort(array, begin, end):
        if begin >= end:
            return
        pivot = q_partion(array, begin, end)
        _quicksort(array, begin, pivot-1)
        _quicksort(array, pivot+1, end)
        return array
    return _quicksort(array, begin, end)


def q_partion(ary,left,right):
    """
    left    左指针位置
    right   右指针位置
    return  返回游标，既确定排序的数的指针，也是左右指针相会处。
    """
    #快排函数，ary为待排序数组，left为待排序的左边界，right为右边界
    pivot = ary[left]  #取最左边的为基准数
    lp = left         #左指针
    rp = right          #右指针
    while lp < rp :  #如果左指针比右指针小，一直循环
        # 右边的指针先移动，寻找比基准值更小的值，找到了或超过了左指针停止        # 为什么从 end 开始移动指针？
        # 如果先从 begin 出发，左指针发现了更大的数，但右指针没发现更小的数。
        # 左右指针相会在更大的数，所以做莫名其妙的置换 begin 和 更大的左指针数
        # 如果从 end 出发，左右指针要么相会在更大的数，要么就在 begin
        # 不会有这种错误。
        while ary[rp] >= pivot and lp < rp : 
            rp -= 1
        # 左边的指针移动，寻找比基准值更大的值，找到了或超过了右指针停止
        while ary[lp] <= pivot and lp < rp :
            lp += 1
        ary[lp],ary[rp] = ary[rp],ary[lp]
    ary[left],ary[lp] = ary[lp],ary[left]
    
    return lp

array = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
quick_sort(array)
```

```python
def partition(array, begin, end):
    pivot = begin
    for i in range(begin+1, end+1):
        if array[i] <= array[begin]:
            pivot += 1
            array[i], array[pivot] = array[pivot], array[i]
    array[pivot], array[begin] = array[begin], array[pivot]
    return pivot


def quicksort(array, begin=0, end=None):
    if end is None:
        end = len(array) - 1
    def _quicksort(array, begin, end):
        if begin >= end:
            return
        pivot = partition(array, begin, end)
        _quicksort(array, begin, pivot-1)
        _quicksort(array, pivot+1, end)
        return array
    return _quicksort(array, begin, end)

array = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
quicksort(array)
```

## 基数排序

```python
import math

def radix_sort(lst, radix=10):
    """a为整数列表， radix为基数"""
    K = int(math.ceil(math.log(max(lst), radix))) # 用K位数可表示任意整数
    bucket = [[] for i in range(radix)] # 不能用 [[]]*radix
    for i in range(1, K+1): # K次循环
        for val in lst:
            bucket[val//(radix**(i-1))%(radix**i)].append(val) # 析取整数第K位数字 （从低到高）
        del lst[:]
        for each in bucket:
            lst.extend(each) # 桶合并
        bucket = [[] for i in range(radix)]
        return lst

array = [50,3,44,38,5,47,15,36,19,48,26,27,2,46,4]
radix_sort(array)
```

## 九种算法效率对比
（共十组代码，两个快速排序）

```python
import random

array = list(range(32,1000,13))
random.shuffle(array)
print('Array Length:', len(array))
print('Array Value:', array)
```

```python
sort_func_list = [
    bubble_sort,
    slct_sort,
    insertion_sort,
    shell_sort,
    heap_sort,
    merge_sort,
    quick_sort,
    quicksort,
    radix_sort,
    sorted,
]
```

```python
import timeit
from collections import OrderedDict

def sort_compare(array,number=10000):
    result = []

    for func in sort_func_list:
        print(func.__name__, end=':')
        t = timeit.timeit(
            '{}(array)'.format(func.__name__), 
            # setup="try:\n    from __main__ import {} \nexcept:\n    from __builtins__ import sorted\nfrom __main__ import array".format(func.__name__),
            globals=globals(),
            number = number
        )
        t_last = t  # 内置算法的时间
        print(t)
        result.append([func.__name__, t])
    print('\n\n')
    print('{:^} {:^12} {:<2} {:^4}'.format('编号','算法名','花费时间','内置模块比较'))
    for num,(key,value) in enumerate(sorted(result,key=lambda x:x[1])):
        print('{:^3} {:<15} {:<.5f} {:<.1%}'.format(num+1,key,value,t/value))
```

```python
sort_compare(array,1000)
```
